```python
# Import the necessary packages
import cv2
import numpy as np
from matplotlib import pyplot as plt
     
```

### 1) Create a blank black image


```python
img = np.ones((600, 600))
plt.imshow(img, cmap='gray')
```

### 2) Create the text using cv2.putText


```python
def load():
    cv2.putText(img, text='PRADEEP', org=(35, 340), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=4, color=(255, 255, 255), thickness=20)

def display_img():
    plt.imshow(img,cmap='gray')
```

```python
load()
display_img()
```

```python
# Create the structuring element
kernal = np.ones((3,3))
```

### 3) Use the Opening Operation


```python
white_noise = np.random.randint(0, 2, size = (600, 600))
white_noise
```

```python
white_noise = white_noise * 255
noise_img = white_noise + img
```

```python
plt.imshow(noise_img, cmap = 'gray')
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernal)
plt.imshow(opening, cmap = 'gray')
```

### 4) Use the Closing Operation


```python
close = cv2.morphologyEx(noise_img, cv2.MORPH_CLOSE, kernal)
plt.imshow(close, cmap = 'gray')
```
